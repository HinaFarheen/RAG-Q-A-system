### **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

texts = [doc.page_content for doc in docs]

# Assuming 'docs' is a list of documents in your corpus
vectorizer = TfidfVectorizer(
    max_df=0.85,          # Ignore terms that appear in more than 85% of documents
    min_df=2,             # Ignore terms that appear in fewer than 2 documents
    max_features=5000,    # Limit vocabulary to the top 5000 terms
    ngram_range=(1, 2),   # Include both unigrams and bigrams
    stop_words='english', # Remove common English stop words
    norm='l2',            # Use L2 normalization (default)
    sublinear_tf=True     # Apply sublinear scaling to term frequencies
)
tfidf_matrix = vectorizer.fit_transform(texts)

def retrieve_document(query):
    # Transform the query into a TF-IDF vector
    query_vec = vectorizer.transform([query])

    # Compute cosine similarity between the query and the corpus
    similarity_scores = cosine_similarity(query_vec, tfidf_matrix).flatten()

    # Retrieve the most relevant document
    top_doc_index = np.argmax(similarity_scores)
    return docs[top_doc_index]
```

```python
def generate_answer(query):
    # Step 1: Retrieve relevant document
    relevant_doc = retrieve_document(query)
    if not isinstance(relevant_doc, str):
        relevant_doc = str(relevant_doc.page_content)

    # Step 3: Generate a response using the LLM
    response = llm(relevant_doc)
    return response
```

```python
import fasttext
import fasttext.util

# Step 1: Download the pre-trained English model if it doesn't already exist
fasttext.util.download_model('en', if_exists='ignore')  # This will download 'cc.en.300.bin'

# Step 2: Load the model
ft = fasttext.load_model('cc.en.300.bin')
```

```python
import numpy as np
# Step 3: Define a function to obtain fasttext embeddings for a document
def fasttext_embed(texts):
    # FastText returns a 300-dimensional vector for each word, so we average them
    embeddings = []
    for text in texts:
        words = text.split()  # Split by spaces, or you can use more advanced tokenization
        word_vectors = np.array([ft.get_word_vector(word) for word in words])
        avg_vector = np.mean(word_vectors, axis=0)  # Average the word vectors to get a single vector for the text
        embeddings.append(avg_vector)
    return embeddings

# Step 4: Prepare your corpus (assuming docs is a list of Document objects)
texts = [doc.page_content for doc in docs]  # Extract the text from Document objects

# Step 5: Create embeddings for the documents using the fasttext model
document_embeddings = fasttext_embed(texts)
```

```python
from langchain.embeddings.base import Embeddings
import numpy as np

class FastTextEmbedding(Embeddings):
    def __init__(self, model):
        self.model = model  # Store the FastText model for reuse

    def embed_documents(self, texts):
        return fasttext_embed(texts)  # Use the existing fasttext_embed function

    def embed_query(self, query):
        # For query, we can use the same fasttext model to generate embeddings
        return np.mean([self.model.get_word_vector(word) for word in query.split()], axis=0)
```

```python
from langchain.vectorstores import FAISS

# Instantiate the custom embedding function
embedding_function = FastTextEmbedding(ft)

# Create the FAISS index with the embeddings
vector_store = FAISS.from_texts(texts, embedding_function)
```

```python
import numpy as np
import nltk
from langchain.embeddings.base import Embeddings

# Download nltk tokenizer if not already downloaded
nltk.download('punkt')

# Load GloVe model (assuming you've already done this)
def load_glove_model(glove_file):
    glove_model = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            glove_model[word] = vector
    return glove_model

# Create a custom embedding function using GloVe
class GloVeEmbeddings(Embeddings):
    def __init__(self, glove_file):
        self.glove_model = load_glove_model(glove_file)

    def embed_documents(self, texts):
        return [self.get_average_vector(text) for text in texts]

    def embed_query(self, query):
        return self.get_average_vector(query)

    def get_average_vector(self, text):
        words = nltk.word_tokenize(text.lower())  # Tokenize and convert to lowercase
        vectors = [self.glove_model[word] for word in words if word in self.glove_model]
        if vectors:
            return np.mean(vectors, axis=0)
        else:
            return np.zeros(100)  # Return a zero vector if no words found in GloVe

# Load the GloVe model (100D in this case)
glove_embeddings = GloVeEmbeddings('glove.6B.100d.txt')
```

```python
from rank_bm25 import BM25Okapi
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
from langchain.chains import RetrievalQA
from langchain import hub

# Extract text content from each Document
docs_content = [doc.page_content for doc in docs]

# Tokenize each document's content for BM25
tokenized_docs = [doc.split(" ") for doc in docs_content]

# Initialize BM25 with tokenized documents
bm25 = BM25Okapi(tokenized_docs)

def bm25_retrieve(query, k=5):
    """
    Retrieves the top `k` documents based on BM25 relevance to the query.
    """
    query_tokens = query.split(" ")
    doc_scores = bm25.get_scores(query_tokens)
    top_k_indices = sorted(range(len(doc_scores)), key=lambda i: doc_scores[i], reverse=True)[:k]
    return [docs_content[i] for i in top_k_indices]

# Formatting the retrieved documents for RAG
def format_docs(docs):
    return "\n\n".join(docs)

# Define questions to ask
questions = [
    "What are the most popular destinations in Pakistan?",
    "Which historical landmarks should I visit in Lahore?",
    "What is the best time to visit northern Pakistan?",
    "What are some must-visit beaches in Pakistan?",
    "Can you suggest a 7-day travel itinerary for exploring Gilgit-Baltistan?",
    "Which cities in Pakistan have the best street food?",
    "What safety tips should female solo travelers keep in mind when visiting Pakistan?",
    "What is the temperature range in Naran during the summer season?"
]

# Example function to ask the questions and retrieve answers
def ask_questions(rag_chain, questions):
    for question in questions:
        top_docs = bm25_retrieve(question)  # Retrieve top documents for each question
        context = format_docs(top_docs)  # Format documents as context for RAG
        response = rag_chain.invoke({"context": context, "question": question})  # Pass context and question to RAG chain
        print(f"Question: {question}")
        print(f"Answer: {response}\n")

# Load the prompt template for RAG (from your hub or repository)
prompt = hub.pull("rlm/rag-prompt")

# Set up the RAG chain
rag_chain = (
    {"context": RunnablePassthrough() | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
)
```

```python
# Execute the function with the RAG chain and list of questions
ask_questions(rag_chain, questions)
```

### **FAISS with GloVe Embedding**

```python
import numpy as np
import nltk
from langchain.embeddings.base import Embeddings

# Download nltk tokenizer if not already downloaded
nltk.download('punkt')

# Load GloVe model (assuming you've already done this)
def load_glove_model(glove_file):
    glove_model = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            glove_model[word] = vector
    return glove_model

# Create a custom embedding function using GloVe
class GloVeEmbeddings(Embeddings):
    def __init__(self, glove_file):
        self.glove_model = load_glove_model(glove_file)

    def embed_documents(self, texts):
        return [self.get_average_vector(text) for text in texts]

    def embed_query(self, query):
        return self.get_average_vector(query)

    def get_average_vector(self, text):
        words = nltk.word_tokenize(text.lower())  # Tokenize and convert to lowercase
        vectors = [self.glove_model[word] for word in words if word in self.glove_model]
        if vectors:
            return np.mean(vectors, axis=0)
        else:
            return np.zeros(100)  # Return a zero vector if no words found in GloVe

# Load the GloVe model (100D in this case)
glove_embeddings = GloVeEmbeddings('glove.6B.100d.txt')
```
